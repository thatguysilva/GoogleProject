```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
col_list = ["Credor", "Total Pago"]
df = pd.read_csv('relatorio(1).csv',  sep=';|;;', engine='python', usecols=col_list, quotechar='.|,')
df['Total Pago'] = df['Total Pago'].str.replace('.','')
df['Total Pago'] = df['Total Pago'].str.replace(',','.')
print(df.to_string())
```

```python
total_pay = df.groupby(by='Credor').agg(
    lambda x: x.astype(float).sum()
)
total_pay

realpay = total_pay.sort_values(by=['Total Pago'], ascending=False)
realpay
```

```python
total_pay.to_csv('total.csv')
```

```python
total_pay.plot(kind='pie', y='Total Pago', legend=False, labels=None)
realpay.iloc[0:10].plot(kind='bar', y='Total Pago')
```

```python
total_pay.plot(kind='bar', y='Total Pago', legend=False, labels=None)
```
